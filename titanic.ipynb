```python
train_df.describe(include = "all")
```

```python
train_df['Male']=np.where(train_df['Sex'] == 'male', 1, 0)

'''
C = Cherbourg (0), Q = Queenstown(1), S = Southampton(2)
'''
train_df['Embarked'].replace(['C','Q','S'], [0,1,2], inplace=True)

train_df['Embarked'].fillna(method = 'ffill', inplace=True)
train_df['Embarked']= train_df['Embarked'].astype('int32')
train_df['FamSize'] = train_df['SibSp']+train_df['Parch']
train_df
```

```python
train_df[['Survived','Age', 'Male', 'Pclass', 'SibSp','Parch','FamSize','Embarked']].corr()
```

```python
age = train_df[['Pclass', 'Age']].groupby('Pclass').mean()
age
```

```python
train_df['Age'] = train_df['Age'].fillna(train_df[['Pclass', 'Age']].groupby('Pclass')['Age'].transform('mean'))
train_df
```
